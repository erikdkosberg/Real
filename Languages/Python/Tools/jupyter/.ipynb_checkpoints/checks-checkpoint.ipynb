```python
#Global Script
"""
This is the global script:

Here you can store variables, functions, and any other scripts you'd like.
The advantage of storing in a global location is that it allows you to
reference these pieces of information at any point you would like in the script.

Stored info should be kept to a consistent naming convention across most inspections

List of important variables, functions, scripts, etc. stored here:
    lineNum
    itemNum
    numRows
    numFlats
    firstFlat
    cavList
    maxCav
    TotalCavs
    count
    partnoBuilder() --> single digit and double digit
    partnoBuilder_DMS()
    partnoBuilder_QA()
    serial_port()
    serial_read()
    serial_close()
"""

Var1 = login.username

if "BT" in login.username:
    lineNum = Var1[:1]
else:
    lineNum = Var1[-2:]

#lineNum = "Erik"

def query_itemNum(lineNum):
    q1 = "Select ItemNum from tbl_GLOBAL_ACTIVEItems "
    q2 = "Where Line = '" + lineNum + " ' "
    Q = q1 + q2
    a = sql.getcolumn(Q, "ItemNum")
    return a
    
ListSplit = query_itemNum(lineNum)    
itemNum = ListSplit[0]

def query_numRows(lineNum):
    q12 = "Select NumRows from tbl_GLOBAL_ACTIVEItems "
    q23 = "Where Line = '" + lineNum + " ' "
    Q = q12 + q23
    b = sql.getcolumn(Q, "NumRows")
    return b
    
ListSplit2 = query_numRows(lineNum)    
numRows = ListSplit2[0]

def query_numFlats(lineNum):
    q8 = "Select NumFlats from tbl_GLOBAL_ACTIVEItems "
    q9 = "Where Line = '" + lineNum + " ' "
    Q = q8 + q9
    b = sql.getcolumn(Q, "NumFlats")
    return b
    
ListSplit3 = query_numFlats(lineNum)    
numFlats = ListSplit3[0]

def query_firstFlat(lineNum):
    q3 = "Select FirstFlat from tbl_GLOBAL_Sampling "
    q4 = "Where Line = '" + lineNum + " ' "
    Q2 = q3 + q4
    c = sql.getcolumn(Q2, "FirstFlat")
    return c
   
ListSplit4 = query_firstFlat(lineNum)
firstFlat = ListSplit4[0]

"""
Grab the flags from the sampling table in sql, store the string results
in a list, index the list for each respective flag, assign an additional
variable named l_flag to store the integer representation of this flag
"""
def query_flags():
    q3 = "Select * from tbl_GLOBAL_Sampling "
    q3 += "Where Line = '" + lineNum + " ' "
    d = sql.gettable(q3)
    return d 
   
ListSplit5 = query_flags()
flagTable = ListSplit5[0]

WtFlag = flagTable[4]
T1Flag = flagTable[5]
T2Flag = flagTable[6]
T2AFlag = flagTable[7]
T3Flag = flagTable[8]
T4Flag = flagTable[9]
T5Flag = flagTable[10]
T6Flag = flagTable[11]
ShearFlag = flagTable[12]
OutsideDiamFlag = flagTable[13]
InsideDiamFlag = flagTable[14]
InsertDiamFlag = flagTable[15]
LipVarFlag = flagTable[16]

l_WtFlag = int(WtFlag)
l_T1Flag = int(T1Flag)
l_T2Flag = int(T2Flag)
#l_T2AFlag = int(T2AFlag)
l_T3Flag = int(T3Flag)
l_T4Flag = int(T4Flag)
l_T5Flag = int(T5Flag)
l_T6Flag = int(T6Flag)
l_ShearFlag = int(ShearFlag)
l_OutsideDiamFlag = int(OutsideDiamFlag)
l_InsideDiamFlag = int(InsideDiamFlag)
l_InsertDiamFlag = int(InsertDiamFlag)
l_LipVarFlag = int(LipVarFlag)

"""
Create and increment a string of the stored procedure to read and update
the current sampling plan, passing in the flag values;

Define a function to execute the stored procedures for updating the sampling plan
with passable specified parameters

"""

def usp_ALL_update(lineNum, firstFlat, WtFlag, T1Flag, T2Flag, T3Flag, T4Flag, T5Flag, T6Flag, OutsideDiamFlag, InsideDiamFlag, InsertDiamFlag, LipVarFlag, ShearFlag):
    all_update = "EXEC usp_GLOBAL_Sampling_ALL_Update @Line = '" + lineNum + "', @FirstFlat = '" + firstFlat + "', @Product = 'Mars Trays', @WtFlag = '" + WtFlag + "', @T1Flag = '" + T1Flag + "', "
    all_update += "@T2Flag = '" + T2Flag + "', @T3Flag = '" + T3Flag + "', @T4Flag = '" + T4Flag + "', @T5Flag = '" + T5Flag + "', @T6Flag = '" + T6Flag + "',"
    all_update += "@OutsideDiamFlag = '" + OutsideDiamFlag + "', @InsideDiamFlag = '" + InsideDiamFlag + "', @InsertDiamFlag = '" + InsertDiamFlag + "', @LipVarFlag = '" + LipVarFlag + "', @ShearFlag = '" + ShearFlag + "'"
    sql.execute(all_update)
    
numRows = int(numRows)
numFlats = int(numFlats)

"""
The following functions will be used to build part numbers for various numeric inputs.
Because GainSeeker cuts off leading zeroes for integer variables and there is a predefined amount
of characters in each part number, we have to account for the portion of the part number where a single
digit value takes up less space than a double digit value. For example, we are building our part numbers
based off the number of rows, and if we have 8 rows on a part number, that will have one less character
than a part number with 12 rows. 

To counteract this, I made two separate functions:
    partno_builder_single
    partno_builder_double
    
The 'single' version of this function hard codes a zero to replace the leading zero that gets cut off
whereas the 'double' version simply passes the value.

Note: the functions listed without _single or _double for partno builders are not intended to be in use, but
have not been deleted in case they may have been referenced in another place.

"""
def partno_builder(checkName):
    p = str(itemNum)
    p += "  ("
    p += str(numRows)
    p += "rows)  "
    p += checkName
    return p
    
def partno_builder_single(checkName):
    p = str(itemNum)
    p += "  (0"
    p += str(numRows)
    p += "rows)  "
    p += checkName
    return p
    
def partno_builder_double(checkName):
    p = str(itemNum)
    p += "  ("
    p += str(numRows)
    p += "rows)  "
    p += checkName
    return p
    
def partno_builder_DMS(defectlistlabel):
    j = "Line "
    j += str(lineNum)
    j += " "
    j += defectlistlabel
    return j
    
#pull check names from ni label
#checkName = [Break Strength, NOT Broken, Broken Detents, Broken Plt]
#1234567  checkName  

def partno_builder_BS(checkName):
    p = str(itemNum)
    p += "  "
    p += checkName
    return p
    
#Grab the last four characters of the inspection name
#This should be the type of check used for QA specific checks (i.e. 1row, 5flt, pick, rchk, etc.)
x = inspect.name
checkType = x[-4:]
checkTypeO = x[-7:]

def partno_builder_QA(checkName):
    p = str(itemNum)
    p += "  ("
    p += str(numRows)
    p += "rows)  "
    p += checkType
    p += "  "
    p += checkName
    return p
    
def partno_builder_QA_single(checkName):
    p = str(itemNum)
    p += "  (0"
    p += str(numRows)
    p += "rows)  "
    p += checkType
    p += "  "
    p += checkName
    return p
    
def partno_builder_QA_double(checkName):
    p = str(itemNum)
    p += "  ("
    p += str(numRows)
    p += "rows)  "
    p += checkType
    p += "  "
    p += checkName
    return p
    
def partno_builder_Opacity_single(checkTypeO):
    p = str(itemNum)
    p += "  (0"
    p += str(numRows)
    p += "rows)  "
    p += checkTypeO
    return p
    
def partno_builder_Opacity_double(checkTypeO):
    p = str(itemNum)
    p += "  ("
    p += str(numRows)
    p += "rows)  "
    p += checkTypeO
    return p
    
"""
three functions to define port settings, read from the buffer, and close the port
(serial_port, serial_read, serial_close)

Note: write commands have been included in the read function for brevity purposes
"""
def serial_port(x):
    serial[0].portnumber = x
    serial[0].baudrate = 9600
    serial[0].parity = "N"
    serial[0].databits = 8
    serial[0].stopbits = 1
    serial[0].handshake = "None"
    serial[0].eolstring = "\r"
    serial[0].filemode = False
    serial[0].open()
    
def serial_read_1(a):
    serial[0].write("<R" + a + "")
    strInput0 = serial[0].readline()
    #strInput0 = serial[0].readcolumn(0,',',True,False)
    com_list0 = []
    i_str = float(strInput0)
    com_list0.append(i_str)
    return com_list0
    
def serial_read_2(a,b):
    serial[0].write("<R" + a + "")
    serial[0].write("<R" + b + "")
    strInput0 = serial[0].readcolumn(0,',',True,False)
    strInput1 = serial[0].readcolumn(0,',',True,False)
    com_list0 = []
    i_str = float(strInput0)
    com_list0.append(i_str)
    com_list1 = []
    i_str = float(strInput1)
    com_list1.append(i_str)
    return com_list0, com_list1
    
def serial_read_4(a,b,c,d):
    serial[0].write("<R" + a + "")
    serial[0].write("<R" + b + "")
    serial[0].write("<R" + c + "")
    serial[0].write("<R" + d + "")
    strInput0 = serial[0].readcolumn(0,',',True,False)
    strInput1 = serial[0].readcolumn(0,',',True,False)
    strInput2 = serial[0].readcolumn(0,',',True,False)
    strInput3 = serial[0].readcolumn(0,',',True,False)
    com_list0 = []
    i_str = float(strInput0)
    com_list0.append(i_str)
    com_list1 = []
    i_str = float(strInput1)
    com_list1.append(i_str)
    com_list2 = []
    i_str = float(strInput2)
    com_list2.append(i_str)
    com_list3 = []
    i_str = float(strInput3)
    com_list3.append(i_str)
    return com_list0, com_list1, com_list2, com_list3
    
def serial_close():
    serial[0].close()
    
#All done, so close the serial port. This is not really needed as all serial ports get closed when the script finishes.

cavPass = 1

##### Un-comment whichever sampling plan you would like to use    
#misc.executescript("CavLoop: 1 part per row")
#misc.executescript("CavLoop: 2 parts per row")
#misc.executescript("CavLoop: Five Flat")
#misc.executescript("CavLoop: Pick")
#misc.executescript("CavLoop: One Row")
#misc.executescript("CavLoop: Recheck")

count = 0
```

```python
#Shift-Script

from datetime import datetime, timedelta
from sets import Set

"""
This script will generate the current shift being run at the time of the inspection

It finds out what day is on the cycle out of 14 and filters based on the time of day

The shift sequence is as follows(Starting July 1, 2020)
    AB x2
    CD x3
    AB x2
    CD x2
    AB x3
    CD x2

"""
#Initialize master start time to compare against and grab the current time
Master_Start_Time = datetime(month=7, day=5, year=2020, hour=23, minute=59, second=59)
Live_Time = datetime.now()

#Subtract master time from current time to get the difference, given in full date format, then take the days
Diff_Days = (Live_Time - Master_Start_Time) 
DayConvert = Diff_Days.days 

#The days since the master data modulo the days in the cycle tells you what day you're on, add one to fix one off
outpu = DayConvert % 14
output = outpu + 1 

#Current hour to filter either A/B or C/D; you can add values here to test
CurrentHour = hsidate.hour

#Initialize a set telling you which days belong to each shift
AB = {1,2,5,6,7,10,11}
CD = {3,4,8,9,12,13,14}

#Still the shift from the night before when its earlier than 8am
if CurrentHour < 8:
    output = outpu
    if output == 0:
        output = 14
        
#Conditional for the output to find what set it's in; lastly we filter by time of day to find the current shift
if output in AB:

    if 8 <= CurrentHour < 20:
    
        inspect.cursubi.trace("shiftID").value= "A"
        inspect.cursubi.trace("shiftID").enabled = False
        
    else:
    
        inspect.cursubi.trace("shiftID").value= "B"
        inspect.cursubi.trace("shiftID").enabled = False
        
elif output in CD:

    if 8 <= CurrentHour < 20: 

        inspect.cursubi.trace("shiftID").value= "C"
        inspect.cursubi.trace("shiftID").enabled = False
    
    else:

        inspect.cursubi.trace("shiftID").value= "D"
        inspect.cursubi.trace("shiftID").enabled = False

```
